```python
import os 
import cv2

```

```python
source_img_dir = './11_8/'

img1_name = '1_deblur_5_4.png'
img2_name = '1_deblur_5_5.png'

img1 = cv2.imread(source_img_dir + img1_name)
img2 = cv2.imread(source_img_dir + img2_name)

img3 = img1 * 0.0 + img2 * 1

cv2.imwrite(source_img_dir+ '1_deblur_R2KAC.png', img3)
```
