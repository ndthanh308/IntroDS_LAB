```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Roll curve parameters
t = np.linspace(0, 12, 100)
x = ((1.1**t) * np.cos(t) + 2) / 4
y = ((1.1**t) * np.sin(t) + 2) / 4
z = np.ones_like(t)

# Arrow parameters for the first plot
arrow_t = 12*np.linspace(0, 1, 100)
arrow_y = ((1.1**arrow_t) * np.cos(arrow_t) + 2) / 4
arrow_z = ((1.1**arrow_t) * np.sin(arrow_t) + 2) / 4

# Arrow parameters for the second plot (empty for now)
arrow_t2 = np.linspace(0, 1, 100)
arrow_t2 = 12*np.rint(arrow_t2*10)/10 
arrow_y2 = ((1.1**arrow_t2) * np.cos(arrow_t2) + 2) / 4
arrow_z2 = ((1.1**arrow_t2) * np.sin(arrow_t2) + 2) / 4


# Arrow parameters for the second plot (empty for now)
arrow_t3 = np.linspace(0, 1, 100)
arrow_t3_int = np.floor(arrow_t3*12)
arrow_t3_rat = arrow_t3*12 - arrow_t3_int
arrow_t3 = arrow_t3_int + 2*arrow_t3_rat - arrow_t3_rat**2  
arrow_y3 = ((1.1**arrow_t3) * np.cos(arrow_t3) + 2) / 4
arrow_z3 = ((1.1**arrow_t3) * np.sin(arrow_t3) + 2) / 4

# Create a 3D plot
fig = plt.figure()
ax1 = fig.add_subplot(111, projection='3d')

# Plot the roll curve in the first subplot
ax1.plot(x, y, z, color='darkblue')

# Add arrows to the first subplot
for i in range(len(arrow_t)):
    ax1.quiver(arrow_t[i]/12, 0, 0,
               arrow_y[i]-arrow_t[i]/12, arrow_z[i], 1,
               color='green', arrow_length_ratio=0.1,
               linewidth = 0.5)

    
# Set limits and labels for the first subplot

ax1.set_xlim(-0.2, 1.2)
ax1.set_ylim(-0.7, 0.7)
ax1.set_zlim(0, 1.5)
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')
ax1.set_zlabel('Z-axis')

plt.savefig('roll1st.pdf', format='pdf', dpi=300)
# Show the figure
plt.show()

# Create a 3D plot
fig = plt.figure()
ax2 = fig.add_subplot(111, projection='3d')

# Plot the roll curve in the second subplot
ax2.plot(x, y, z, color='darkblue')

# Add arrows to the second subplot
for i in range(len(arrow_t2)):
    ax2.quiver(arrow_t[i]/12, 0, 0,
               arrow_y2[i]-arrow_t[i]/12, arrow_z2[i], 1,
               color='green', arrow_length_ratio=0.1,
               linewidth = 0.5)
# Set limits and labels for the 2d subplot

ax2.set_xlim(-0.2, 1.2)
ax2.set_ylim(-0.7, 0.7)
ax2.set_zlim(0, 1.5)
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y-axis')
ax2.set_zlabel('Z-axis')


plt.savefig('roll2nd.pdf', format='pdf', dpi=300)
# Show the figure
plt.show()


# Create a 3D plot
fig = plt.figure()
ax3 = fig.add_subplot(111, projection='3d')

# Plot the roll curve in the third subplot
ax3.plot(x, y, z, color='darkblue')

# Add arrows to the third subplot
for i in range(len(arrow_t2)):
    ax3.quiver(arrow_t[i]/12, 0, 0,
               arrow_y3[i]-arrow_t[i]/12, arrow_z3[i], 1,
               color='green', arrow_length_ratio=0.1,
               linewidth = 0.5)


# Set limits and labels for the 3d subplot

ax3.set_xlim(-0.2, 1.2)
ax3.set_ylim(-0.7, 0.7)
ax3.set_zlim(0, 1.5)
ax3.set_xlabel('X-axis')
ax3.set_ylabel('Y-axis')
ax3.set_zlabel('Z-axis')

plt.savefig('roll3d.pdf', format='pdf', dpi=300)
# Show the figure
plt.show()

```
